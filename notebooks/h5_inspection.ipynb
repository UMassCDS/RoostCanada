# Print all root level object names (keys)
#### these can be group or dataset names

```python
with h5py.File(filename, "r") as f:
    
    print("Keys: %s" % f.keys())

    for key in f.keys():
        item = f[key]
        print(f"Item: {key}")
        print(f"Type: {type(item)}")

        if isinstance(item, h5py.Group):
            print(f"Keys in group {key}: {list(item.keys())}")
        elif isinstance(item, h5py.Dataset):
            print(f"Shape of dataset {key}: {item.shape}")
```

# Go inside dataset1

```python
with h5py.File(filename, "r") as f:
    # Print keys in 'dataset1'
    print("Keys in 'dataset1': %s" % f['dataset1'].keys())

    for key in f['dataset1'].keys():
        item = f['dataset1'][key]
        print(f"Item: {key}")
        print(f"Type: {type(item)}")

        if isinstance(item, h5py.Group):
            print(f"Keys in group {key}: {list(item.keys())}")
        elif isinstance(item, h5py.Dataset):
            print(f"Shape of dataset {key}: {item.shape}")
```

# get the acutal data (?)

```python
with h5py.File(filename, "r") as f:
    # Iterating through the root groups
    for key in f.keys():
        print(f"Keys in '{key}': {f[key].keys()}")
        group = f[key]
        
        # Iterating through the subgroups
        for subgroup_key in group.keys():
            print(f"Item: {subgroup_key}")
            print(f"Type: {type(group[subgroup_key])}")
            
            subgroup = group[subgroup_key]
            # Checking if the subgroup is an HDF5 group
            if isinstance(subgroup, h5py.Group):
                print(f"Keys in group {subgroup_key}: {subgroup.keys()}")
                
                # Iterating through the dataset in the subgroups
                for dataset_key in subgroup.keys():
                    print(f"Item: {dataset_key}")
                    print(f"Type: {type(subgroup[dataset_key])}")
                    
                    dataset = subgroup[dataset_key]
                    # Checking if the object is an HDF5 dataset
                    if isinstance(dataset, h5py.Dataset):
                        print(f"Values: {dataset[:1]}")  # prints the actual data
```
